```python
"""
Functions vs Method
"""

def my_function(x):
    return x+1

y = 3
result = my_function(y)
print(result)

class my_class:
    @staticmethod
    def my_method(x):
        return x+1

cls = my_class()
cls.my_method(5)
```

```python
"""
Function Parameters
"""

def introduce_myself(name, birthyear, city): #parameter
    print("Hello everyone, My name is %s, I was born in %d, and I live in %s." \
          % (name, birthyear, city))
    
introduce_myself("Hyeon Woo", 1997, "Changwon") #argument
```

```python
"""
Return Keyword
"""

# Leap year = 윤년 (2월 29일)
def check_leap_year(year):
    if year % 4 == 0:
        return "%s is a leap year." % (year)
    else:
        return "%s is not a leap year." % (year)
    
check_leap_year(2021)
```

```python
"""
Returning Multiple Values
"""

def square_point(x,y,z):
    x_squared = x ** 2
    y_squared = y ** 2
    z_squared = z ** 2
    return x_squared, y_squared, z_squared

three_squared, four_squared, five_squared = square_point(3,4,5)

print(three_squared, four_squared, five_squared)
```

```python
"""
Global Variables
"""

a = "Hello"

def prints_a():
    print(a)
    
print(a)
prints_a()
```

```python
"""
Local Variables
"""

b = 2 # global var

def f1():
    b = 2 # local var.
    print(b)
    
f1()
print(b)
```

```python
"""
Keyword Arguments and Default Values
"""

def findvolume(height, width, depth):
    print("Height = %s" % (height))
    print("Width = %s" % (width))
    print("Depth = %s" % (depth))
    
# ex: height = 1, width = 2, depth = 3
print(findvolume(1, 2, 3))
```
